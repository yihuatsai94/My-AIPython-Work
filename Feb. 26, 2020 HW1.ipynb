```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from ipywidgets import interact
```

```python
def myplot(n=1):
    x = np.linspace(-15, 25, 500)
    y = np.sinc(n*x)
    plt.plot(x,y)
```

```python
myplot()
```

```python
interact(myplot, n=(7, 77));
```
