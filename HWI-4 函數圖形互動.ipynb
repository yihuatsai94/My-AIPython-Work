```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from ipywidgets import interact
```

```python
def myplot(n=1):
    x = np.linspace(-20, 10, 350)
    y = np.sinc(n*x)
    plt.plot(x,y)
```

```python
myplot()
```

```python
interact(myplot, n=(77, 77))
```
